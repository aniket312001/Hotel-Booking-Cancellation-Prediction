# Handling Missing data

```python
feature_with_na = [feat for feat in df.columns if df[feat].isnull().sum()>=1]

for feat in feature_with_na:
    print(f'{feat} -> {np.round(df[feat].isnull().mean(),4)}')
```

```python
# missing data
plt.figure(figsize=(13,7))
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
 df.fillna({'children':df['children'].median(),'agent':df['agent'].median(),'country':'missing'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('company',axis=1,inplace=True)
```

```python
df.head(3)
```

# Handling Rare Categorical Feature

We will remove categorical variable that are present less than 1% of the observation

```python
categories_feature = [feat for feat in df.columns if df[feat].dtype=='O']

for feat in categories_feature:
    print(f'No. of Feature in {feat} is {len(df[feat].unique())}')
```

```python
# Imp

for feat in categories_feature:
    temp = df.groupby(feat)['is_canceled'].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[feat] = np.where(df[feat].isin(temp_df),df[feat],'Rare_var')
```

```python
df.head()
```

```python
for feat in categories_feature:
    print(f'No. of Feature in {feat} is {len(df[feat].unique())}')
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

for feat in categories_feature:
    df[feat] = label.fit_transform(df[feat])
```

```python
df.head()
```

```python
df.info()
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

X = df.drop('is_canceled',axis=1)
y = df['is_canceled']

scale = StandardScaler()
scale_feat = scale.fit_transform(X)

new_df = pd.DataFrame(scale_feat,columns=X.columns)
```

```python
new_df = pd.concat([new_df,y],axis=1)
new_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = new_df.drop('is_canceled',axis=1)
y = df['is_canceled']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

lg = LogisticRegression()
lg.fit(X_train,y_train)
lg_predict = lg.predict(X_test)

print(confusion_matrix(y_test,lg_predict))
print()
print(classification_report(y_test,lg_predict))
print()
print(accuracy_score(y_test,lg_predict))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_predict = rf.predict(X_test)

print(confusion_matrix(y_test,rf_predict))
print()
print(classification_report(y_test,rf_predict))
print()
print(accuracy_score(y_test,rf_predict))
```

# Hence Both Logistic Regression & RandomForest perform Best
